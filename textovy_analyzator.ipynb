```python
# Příklad dictionary slov
words_dict = {
    "Apple": "jablko",
    "banana": "banán",
    "Cherry": "třešeň",
    "apricot": "meruňka",
    "Blueberry": "borůvka",
    "Avocado": "avokádo"
}

def count_words_starting_with_capital(words):
    count = 0
    for word in words:
        if word[0].isupper():
            count += 1
    return count

# Výpočet počtu slov, která začínají velkým písmenem
count = count_words_starting_with_capital(words_dict)

print(f'Počet slov začínajících velkým písmenem: {count}')
```

```python
# Příklad textu uloženého v proměnné TEXT jako list
TEXT = ["Ahoj, jak se máš?", "Dnes je krásný Den.", "Python je skvělý Programovací jazyk.", "Ráno jsem šel do Obchodu."]

# Inicializace počítadla
pocet_slov_velkym_pismenem = 0

# Procházení každé věty v TEXT
for veta in TEXT:
    # Rozdělení věty na slova
    slova = veta.split()
    # Kontrola každého slova
    for slovo in slova:
        # Pokud slovo začíná velkým písmenem, zvýší se počítadlo
        if slovo[0].isupper():
            pocet_slov_velkym_pismenem += 1

# Výpis výsledku
print("Počet slov začínajících velkým písmenem:", pocet_slov_velkym_pismenem)

## ověření uživatelů, verze B při více atributech
users = {
    "user1": {"password": "password1", "email": "user1@example.com"},
    "user2": {"password": "password2", "email": "user2@example.com"},
    "user3": {"password": "password3", "email": "user3@example.com"},
    "user4": {"password": "password4", "email": "user4@example.com"}
}

# Proměnná pro uložení stavu přihlášení
login = False

# Získání uživatelského jména
username = input("Zadejte uživatelské jméno: ")

# Kontrola, zda uživatelské jméno existuje ve slovníku
if username in users:
    # Získání hesla
    password = input("Zadejte heslo: ")
    
    # Porovnání zadaného hesla s heslem uloženým ve slovníku
    if users[username]["password"] == password:
        login = True
        print("Přihlášení bylo úspěšné.")
    else:
        print("Nesprávné heslo.")
else:
    print("Uživatelské jméno neexistuje.")

# Výpis výsledku
print("Login:", login)
```
