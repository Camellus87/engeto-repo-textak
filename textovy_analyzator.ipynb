```python
## projekt z engeta, finále v1.0
'''
textovy_analyzator: první projekt pro Engeto Python; v1.0

author: Kamil Hnátek
email: kamil.hnatek@gmail.com
discord: irenicus87
'''
users = {
    "bob": "123",
    "ann": "pass123",
    "mike": "password123",
    "liz": "pass123"
}

login = False

# Ziska login
username = input("Zadejte uživatelské jméno: ")

# Kontrola, zda uživatelské jméno existuje ve slovníku
# má se opakovat dotazování na login? V tom případě změna if na while.
if username in users:
    # Získání hesla
    password = input("Zadejte heslo: ")
    # Kontrola hesla
    if users[username] == password:
        login = True
        print(f"Přihlášení bylo úspěšné. Vítej {username}.")
    else:
        print("Nesprávné heslo. Zkus to znovu.")
else:
    print("Uživatel neexistuje.")

TEXTS = ['''
Situated about 10 miles west of Kemmerer,
Fossil Butte is a ruggedly impressive
topographic feature that rises sharply
some 1000 feet above Twin Creek Valley
to an elevation of more than 7500 feet
above sea level. The butte is located just
north of US 30N and the Union Pacific Railroad,
which traverse the valley. ''',
'''At the base of Fossil Butte are the bright
red, purple, yellow and gray beds of the Wasatch
Formation. Eroded portions of these horizontal
beds slope gradually upward from the valley floor
and steepen abruptly. Overlying them and extending
to the top of the butte are the much steeper
buff-to-white beds of the Green River Formation,
which are about 300 feet thick.''',
'''The monument contains 8198 acres and protects
a portion of the largest deposit of freshwater fish
fossils in the world. The richest fossil fish deposits
are found in multiple limestone layers, which lie some
100 feet below the top of the butte. The fossils
represent several varieties of perch, as well as
other freshwater genera and herring similar to those
in modern oceans. Other fish such as paddlefish,
garpike and stingray are also present.'''
]

prvni_text = TEXTS[0]
druhy_text = TEXTS[1]
treti_text = TEXTS[2]

if login is True:
    print("Máme 3 texty k analýze.")
    print("----------------------------------------")
    cislo_textu=input('Jaký text chceš analyzovat? 1, 2 nebo 3:')
    cislo_x = 0
    print(f"Zvolil jsi text číslo: {cislo_textu}")
    print("----------------------------------------")

    # Kontrola vstupu uzivatele, jaký text si zvolil
    if cislo_textu == '1':
        vybrany_text = prvni_text
        cislo_x = 0
    elif cislo_textu == '2':
        vybrany_text = druhy_text
        cislo_x = 1
    elif cislo_textu == '3':
        vybrany_text = treti_text
        cislo_x = 2
    elif cislo_textu != '1' and cislo_textu != '2' and cislo_textu != '3':
        print("Takovy text neexistuje, zkus to znovu.")

    if cislo_textu == '1' or cislo_textu == '2' or cislo_textu == '3':
        # Rozdělení textu na slova
        slova = TEXTS[cislo_x].split()

        # Inicializace počítadla
        pocet_slov_velkym_pismenem = 0
        pocet_slov_malym_pismenem = 0
        pocet_cisel = 0
        pocet_slov = 0
        cisla_list = []

        #Kontrola každého slova a vypocet poctu slov bez cislic
        for slovo in slova:
           if not slovo.isdigit():
               pocet_slov += 1

        # print(pocet_slov)
        # Kontrola každého slova
        for slovo in slova:
            # Pokud slovo začíná velkým písmenem, zvýší se počítadlo
            if slovo[0].isupper():
                pocet_slov_velkym_pismenem += 1
            if slovo[0].islower():
                pocet_slov_malym_pismenem += 1
            if slovo.isdigit():
                pocet_cisel += 1
                cisla_list.append(int(slovo))
        
        # Převede ciselny list na integer a udělá sumu
        int_numbers = [int(num) for num in cisla_list]
        total_sum = sum(int_numbers)

        # Spočítání délek slov
        word_lengths = [len(word) for word in slova]

        # Spočítání četnosti jednotlivých délek slov pomocí slovníku
        length_counts = {}
        for length in word_lengths:
            if length in length_counts:
                length_counts[length] += 1
            else:
                length_counts[length] = 1

        # Seřazení dle délek slov
        sorted_lengths = sorted(length_counts.items())

        # Výpis výsledku
        print("Počet slov začínajících velkým písmenem:", pocet_slov_velkym_pismenem)
        print("Počet slov začínajících malým písmenem:", pocet_slov_malym_pismenem)
        print("Počet čísel:", pocet_cisel)
        print("Suma čísel:", total_sum)
         # Vytvoření a zobrazení sloupcového grafu
        print("----------------------------------------")
        print(f"{'LEN':<3}| {'OCCURENCES':<20}|{'NR.':<3}")
        print("----------------------------------------")
        for length, count in sorted_lengths:
            print(f"{length:<3}| {'*' * count:<20}|{count}")
        
```
